```python
import os
import numpy
import matplotlib.pyplot as plt
import glob
```

```python
path="D:\\si\\Faces for Training\\Faces for Training"
```

```python
dir_paths=os.listdir(path)
```

```python
count=[]
ids={}
for i in dir_paths:
    loc=os.path.join(path,i)
    images_path = glob.glob(os.path.join(loc, "*.*"))
    count.append(len(images_path))
    ids[i]=len(images_path)
    print(f" {i} - {len(images_path)}")
```

```python
stat={}
stats_count={}
```

```python
for i in range(1,max(count)+1):
    c=[]
    if i in count:
        for x in range(len(count)):
            if count[x]==i:
                c.append(dir_paths[x])
        stat[i]=c
        stats_count[i]=len(c)
            
```

```python
fre=list(stats_count.keys())
no=list(stats_count.values())
```

```python
plt.bar(fre, no, color ='maroon')

plt.xlabel("Number of images")
plt.ylabel("Number of ID's")
plt.show()
```

```python
file_path=os.path.join("D:\\si\\Faces for Training","stats.txt")
```

```python
file=open(file_path,"w")
```

```python
for i in fre:
    file.write(f"{i} -- {stat[i]} \n")
file.close()
```
